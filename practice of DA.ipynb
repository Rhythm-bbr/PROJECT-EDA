```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\dilip patel\\Downloads\\Python_Diwali_Sales_Analysis-main\\Diwali Sales Data.csv",encoding = "unicode_escape")
df.head(10)
```

```python
df.drop(["Status", "unnamed1"],axis=1,inplace=True)
df
```

```python
df.loc[:,df.isnull().any(axis=0)]
```

```python
df.loc[df.isnull().any(axis=1),:]
```

```python
indicates = df[df["Amount"].isnull()].index[0:8]
fill = df["Amount"].ffill()
df.loc[indicates,"Amount"] = fill.loc[indicates] 
# this will do ffill to only 8 nan values in Amount without giving a f about how many groups
df.loc[df.isnull().any(axis=1),:]
```

```python
df.bfill(axis=0,inplace=True)
df.isnull().sum()
```

```python
(df["Cust_name"].str.lower() == "joni").sum()
```

```python
df["Cust_name"].str.contains("joni", case=False, na=False).sum()
#This will count similar names like- joNi, JOni, 123_joNi4, etc
```

```python
stt = df.groupby("State",as_index=False)["Gender"].value_counts()
stt.rename(columns={"count":"Total number of M OR F"},inplace=True)
pd.DataFrame(stt)
stt
```

```python
newstt = df.groupby(["State","Gender"],as_index=False)["Amount"].sum()
newstt
```

```python
Amt = df.groupby(["Gender"], as_index=False)["Amount"].sum()
Amt
```

```python
Orders = df.groupby(["Gender"], as_index=False)["Orders"].sum()
Orders
```

```python
sns.set_theme(style="darkgrid",rc={"figure.figsize":(12,10)})

plt.subplot(2,2,1)
ax = sns.barplot(x="Gender",y="Orders",data=Orders,order={"M","F"},palette={"M":"b","F":"pink"},hue="Gender")

for i in ax.containers:
    ax.bar_label(i)

plt.subplot(2,2,2)
axx = sns.barplot(x="Gender",y="Amount",data=Amt,order={"M","F"},palette={"M":"b","F":"pink"},hue="Gender")

for i in axx.containers:
    axx.bar_label(i)

plt.subplot(2,2,3)
ay = sns.histplot(x="Age Group",data=df,discrete=True,hue="Gender",hue_order=["M","F"],palette="cool",fill=True,element="step",multiple="dodge",shrink=0.8)

plt.subplot(2,2,4)
ayy = sns.countplot(x="Marital_Status",data=df,hue="Gender",hue_order=["M","F"],palette="Accent")
plt.xticks([0,1],["Unmarried","Marrid"])

for i in ayy.containers:
    ayy.bar_label(i)


plt.show()
```

```python
sns.set_theme(style="darkgrid",rc={"figure.figsize":(40,20)})

plt.subplot(2,1,1)
sns.lineplot(x="State",y="Total number of M OR F",data=stt,hue="Gender",hue_order=["M","F"],style="Gender",markers="o",markersize=5,markerfacecolor="black",dashes=False)

plt.subplot(2,1,2)
sns.lineplot(x="State",y="Amount",data=newstt,hue="Gender",hue_order=["M","F"],style="Gender",markers="o",markersize=5,markerfacecolor="black",dashes=False)

plt.legend(fontsize="xx-large",shadow=True)

plt.show()
```
